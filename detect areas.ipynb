```python
import cv2
import numpy as np
import pickle
import pandas as pd
from ultralytics import YOLO
import cvzone



my_file = open("coco.txt", "r")
data = my_file.read()
class_list = data.split("\n")
#read the width of yolo
model=YOLO('yolov8s.pt')

with open("computervision",'rb') as f:
                data = pickle.load(f)
                polylines , areaName = data['polylines'] , data['areaName']
            
cap=cv2.VideoCapture('video.mp4')

count=0

while True:
    ret, frame = cap.read()
    if not ret:
        cap.set(cv2.CAP_PROP_POS_FRAMES, 0)
        continue
   
    count += 1
    if count % 3 != 0:
        continue

    frame=cv2.resize(frame,(1200,700))
    frame_copy = frame.copy()
    results=model.predict(frame)
    a=results[0].boxes.data
    px=pd.DataFrame(a).astype("float")
    
    list1=[]
    for index,row in px.iterrows():
 
        x1=int(row[0])
        y1=int(row[1])
        x2=int(row[2])
        y2=int(row[3])
        #index 5 contain class name
        d=int(row[5])
        #return classs name from its index
        c=class_list[d]
        cx=int(x1+x2)//2
        cy=int(y1+y2)//2

        if 'car' in c:
            #save the bonding box (center) of car in list
            list1.append([cx,cy])
            
            #display the bounding box of cars
            #cv2.rectangle(frame,(x1,y1),(x2,y2),(255,0,0),2)
            
    counter1 = []        
    for i, polyline in enumerate (polylines):
        
        
        #display the polylines 
        #cv2.polylines(frame, [polyline], True, (0, 255, 0), 2)
        #display the numbers of polylines
        #cvzone.putTextRect(frame,f'{areaName[i]}' , tuple(polyline[0]),1,1)
        
        for l in list1:
            cx1 = l[0]
            cy1 = l[1]
            result = cv2.pointPolygonTest(polyline,(cx1,cy1),False)
            #detect and counts the area based on the center of detected cars
            if result >=0:
                cv2.circle(frame,(cx1,cy1),5,(0,255,0),-1)
                
                #change the polylines to red after it becomes full
                #cv2.polylines(frame, [polyline], True, (0, 0, 255), 2)
                
                counter1.append(cx1)
     
    car_count = len(counter1)
    
    #if we want to put the displayed text in a rectangle 
    #cvzone.putTextRect(frame,f'busy places: {car_count} ' ,(50,60),1.2,4,  colorT=(0,0,255), colorR=(255, 255, 255))
    #cvzone.putTextRect(frame,f'Empty places: {free_space} ' ,(50,110),1.2,4,colorT=(0,255,0), colorR=(255, 255, 255))

    cv2.putText(frame, f'busy places: {car_count} ', (50,60), cv2.FONT_HERSHEY_SIMPLEX, 1.2, (0,0,255), 4)
    free_space = 6 - car_count
    cv2.putText(frame, f'Empty places: {free_space} ', (50,110), cv2.FONT_HERSHEY_SIMPLEX, 1.2, (0,255,0), 4)
    
    cv2.imshow('FRAME', frame)
    key = cv2.waitKey(1)
    if key == ord('q'):
        break


cap.release()

cv2.destroyAllWindows()

```
